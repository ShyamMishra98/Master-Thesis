````python
# clone the repositories mtm2017-tutorial and moses for Romanian English preprocessing scripts.
# the required datasets are also included in the mtm2017-tutorial
# moses helps in normalizing punctuation, tokenizing during preprocessing and recasing during preprocessing and also postprocessing

%cd ../..

!git clone https://github.com/marian-nmt/mtm2017-tutorial
!git clone https://github.com/marian-nmt/moses-scripts
````

````python
# subword-nmt is used for segmenting texts into subword units in preprocessing

!pip install subword-nmt
````

````python
# pretrained Marian NMT model for RO-EN translation along with their datasets is downloaded

!wget -r -e robots=off -nH -np -R index.html* http://data.statmt.org/wmt16_systems/ro-en/
````

````python
# preprocessing of Romanian and English sentences.

!cat mtm2017-tutorial/test/newstest2016.ro \ | ./moses-scripts/scripts/tokenizer/normalize-punctuation.perl -l ro \ | ./mtm2017-tutorial/scripts/normalise-romanian.py \ | ./mtm2017-tutorial/scripts/remove-diacritics.py \  | ./moses-scripts/scripts/tokenizer/tokenizer.perl -a -l ro \ | ./moses-scripts/scripts/recaser/truecase.perl -model wmt16_systems/ro-en/truecase-model.ro \ |  subword-nmt apply-bpe -c wmt16_systems/ro-en/roen.bpe > wmt16_systems/ro-en/newstest2016.tok.tc.bpe.ro


!cat mtm2017-tutorial/test/newstest2016.en \ | ./moses-scripts/scripts/tokenizer/normalize-punctuation.perl -l en \ | ./moses-scripts/scripts/tokenizer/tokenizer.perl -a -l en > wmt16_systems/ro-en/newstest2016.tok.en
````

**Create a config.ende.yml file with the contents below and upload to this notebook :**

```
#File: config.ende.yml
  ```

````python
# translation of source Romanian sentences to target English sentences using the pretrained model "model.npz" and the configuration file


!cat wmt16_systems/ro-en/newstest2016.tok.tc.bpe.ro  \
  | ./marian-dev/build/marian-decoder -c config.ende.yml \
  > wmt16_systems/ro-en/newstest2016.tok.tc.bpe.en.output
````

````python
# Post processing of the English output machine translations using moses recaser 

!cat wmt16_systems/ro-en/newstest2016.tok.tc.bpe.en.output \
  | sed 's/@@ //g' \
  | ./moses-scripts/scripts/recaser/detruecase.perl \
  > wmt16_systems/ro-en/newstest2016.en.output
````
